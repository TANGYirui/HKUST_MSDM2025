```python
import random

# If n is an odd number,
# the center where the main diagonal crosses the secondary diagonal being 0 may lead to double calculations.
def count_zero(matrix):  
    n = len(matrix)  
    zero_count = 0  
    for i in range(n):  
        # Main diagonal 
        if matrix[i][i] == 0:  
            zero_count += 1  
        # Minor diagonal  
        if matrix[i][n-1-i] == 0:  
            # prevent double count  
            if i != n-1-i:  
                zero_count += 1  
    return zero_count  

if __name__ == "__main__":
    n = int(input("Please input matrix size n (positive integer): "))  
    M = int(input("Please input max random integer M (positive integer): "))  

    if n <= 0 or M < 0:  
        print("n should be positive integer, M should be non-negative integer")  
    else:  
        # generate random matrix 
        matrix = [[random.randint(0, M) for _ in range(n)] for _ in range(n)]  
        # print the matrix 
        print("Random matrix:")  
        for row in matrix:  
            print(row)  
        # count zeros on both diagonals 
        print("Number of zeros on both diagonals:", count_zero(matrix))
```

```python
import math

def quadratic(a, b, c):
    # input validation
    if not (isinstance(a, (int, float)) and isinstance(b, (int, float)) and isinstance(c, (int, float))):
        print("Error: All inputs must be numbers.")
        return
    
    if a == 0:
        print("Error: a cannot be zero (not a quadratic equation).")
        return
    print(f"The quadratic equation is: {a}x^2 + {b}x + {c} = 0")
    
    Delta = b ** 2 - 4 * a * c
    two_a = 2 * a

    def fmt(x):
        # The value should be displayed using at most 3 significant figures
        if int(x) == x:
            return str(int(x))
        else:
            return "{:.3g}".format(x)
    
    if Delta > 0:
        root1 = (-b + math.sqrt(Delta)) / two_a
        root2 = (-b - math.sqrt(Delta)) / two_a
        print(f"The roots are {fmt(root1)} and {fmt(root2)}.")
    elif Delta == 0:
        root = -b / two_a
        print(f"The root is {fmt(root)}.")
    else:
        real = -b / two_a
        imag = math.sqrt(-Delta) / two_a
        # if Imaginary part = 1, do not print 1i, only display i
        real_str = fmt(real)
        imag_str = "" if abs(imag - 1) < 1e-10 else fmt(imag)
        plus = f"{real_str}+{imag_str}i" if imag_str else f"{real_str}+i"
        minus = f"{real_str}-{imag_str}i" if imag_str else f"{real_str}-i"
        print(f"The roots are {plus} and {minus}.")

if __name__ == "__main__":
    try:
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        quadratic(a, b, c)
    except ValueError:
        print("Error: Invalid input. Please enter numeric values for coefficients.")
```

```python
# Part (a)
def hollow_rhombus(a, b, printout = False):
    """
    打印一个最内层菱形边长为 a，边框层数为 b 的空心菱形。
    """
    total_rows = 2 * a + 2 * b - 1  # total lines
    pattern = []
    for i in range(total_rows):
        line = ""

        if i < b:  # 顶部三角形
            # 前导空格
            leading_spaces = " " * (b - i - 1 + a)
            line += leading_spaces
            # 左斜杠
            line += "/" * i
            # 右斜杠
            line += "\\" * i
            # 右边空格
            after_spaces = " " * (b - i - 1 + a)
            line += after_spaces

        elif i < b + a :  # 上半部分平行四边形
            # 前导空格
            leading_spaces = " " * (a + b - i - 1)
            line += leading_spaces
            # 左侧边框 (b 个 '/')
            left_border = "/" * b
            line += left_border
            # 中间空白
            middle_spaces = " " * (2 * (i - b))
            line += middle_spaces
            # 右侧边框 (b 个 '\')
            right_border = "\\" * b # 反斜杠要\\才能作为字符输入
            line += right_border
            # 右边空格
            after_spaces = " " * (a + b - i - 1)
            line += after_spaces

        elif i < b + 2 * a - 1:  # 下半部分平行四边形
            # 前导空格
            leading_spaces = " " * (i - (b + a))
            line += leading_spaces
            # 左侧边框 (b 个 '/')
            left_border = "\\" * b
            line += left_border
            # 中间空白
            middle_spaces = " " * (2 * (b + 2*a - 1 - i))
            line += middle_spaces
            # 右侧边框 (b 个 '\')
            right_border = "/" * b
            line += right_border
            # 右侧空白
            after_spaces = " " * (i - (b + a))
            line += after_spaces

        else:  # 底部三角形
            # 前导空格
            leading_spaces = " " * (i - (b + a))
            line += leading_spaces
            # 左斜杠
            line += "\\" * (total_rows - i)
            # 右斜杠
            line += "/" * (total_rows - i)
            # 右侧空格
            after_spaces = " " * (i - (b + a))
            line += after_spaces

        pattern.append(line)
    if printout :
        for line in pattern:
            print(line)  
    return pattern

# test
lines = hollow_rhombus(5, 3, printout=True)  


def beautiful_pattern():

    while True:
        s = input("Please input rhombus size of the rhombus a, b and n the number of repeatition, use space to seperate, and input q/Q to quit.").strip()
        if s.lower() == 'q':
            print(f"You have input {s} and the program terminates")
            break
        items = s.split()
        if len(items) != 3:
            print("Please input three non-negative integer")
            continue
        if not all(item.isdigit() for item in items):
            print("Every input should be non-negative integer, please input again")
            continue
        a, b, n = map(int, items)
        if a < 0 or b < 0 or n < 0:
            print("Every input should be non-negative integer, please input again")
            continue
        # single rhombus use hollow_rhombus function
        single_rhombus_lines = hollow_rhombus(a, b)
        # n rhombuses
        for line in single_rhombus_lines:
            print((line + " ") * n)

beautiful_pattern()
```

```python
from datetime import date
TODAY = str(date.today())

def my_copyright5(name, email, date=TODAY):
    # 行模板
    line2 = f"***  programmed by {name} for MSDM5002  ***"
    width = len(line2)  # 全文宽度

    # Frame
    star_row = "*" * width

    # Date centered
    date_text = f"date: {date}"
    side_space = width - len(date_text) - 6  # 去除两边***和各2空格
    left_pad = side_space // 2
    right_pad = side_space - left_pad
    line3 = "***" + " " * (2+left_pad) + date_text + " " * (2+right_pad) + "***"
    line3 = line3[:width]  # 若偶数时刚好等宽

    # Dash row
    line4 = "***" + "-" * (width - 6) + "***"

    # 段落正文拆分
    para = 'You can use it as you like, but there might be many bugs. If you find some bugs, please send them to "' + email + '".'
    words = para.split()
    content_lines = []

    current = []
    idx = 0
    while idx < len(words):
        line = '  '  # 每行开头2个空格
        while True:
            word = words[idx]
            # 对email特殊处理（email部分用引号包裹，可能超长，直接当成一个整体再暴力断行即可）
            if '"' in word and "@" in word:
                email_with_quote = word
                # 把引号和邮箱拼起来
                email_full = email_with_quote
                # 填进当前行，不够就截断
                while email_full:
                    leftover = width - 8 - len(line)  # ***+2空格+2空格+*** = 8
                    chunk = email_full[:leftover]
                    line += chunk
                    email_full = email_full[leftover:]
                    # 如果还有剩余就立即结束本行
                    if email_full:
                        line += ""  # 不补空格
                        content_lines.append("***" + line.ljust(width-6) + "***")
                        line = "  "  # 新一行以空格开头
                    else:
                        # 这里只剩最后了，继续while指针
                        if idx + 1 < len(words):
                            if len(line) > 2:
                                line += " "
                        break
                idx += 1
                break
            else:
                # 不是邮件，不能断词
                next_word = word
                extra = ("" if line.endswith("  ") else " ")
                if len(line) + len(extra) + len(next_word) <= width - 6:
                    if not line.endswith("  "):
                        line += " "
                    line += next_word
                    idx += 1
                    if idx == len(words):
                        break
                else:
                    break
        content_lines.append("***" + line.ljust(width-6) + "***")

    # 输出
    print(star_row)
    print(line2)
    print(line3)
    print(line4)
    for l in content_lines:
        print(l)
    print(star_row)

# 示例测试
my_copyright5('IA', 'ia@ust.hk')
print()
my_copyright5('CHAN Tai Man', 'tmchan@asdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdfasdf.com', '2025-01-01')
```
