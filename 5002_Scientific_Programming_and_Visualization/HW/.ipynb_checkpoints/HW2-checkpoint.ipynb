```python
import random

# part(a) 

def check_winner(board):
    """
    Write a function that checks if any player wins..
    Input: board (list of 9 integers, 0=empty, 1=A, -1=B)
    Output: 1 if A wins, -1 if B wins, 0 if no winner yet.
    """
    # Check rows
    for i in range(3):
        if board[3*i] == board[3*i + 1] == board[3*i + 2] != 0:
            return board[3*i]
    # Check columns
    for j in range(3):
        if board[j] == board[3 + j] == board[6 + j] != 0:
            return board[j]
    # Check diagonals
    if board[0] == board[4] == board[8] != 0:
        return board[0]
    if board[2] == board[4] == board[6] != 0:
        return board[2]
    # No winner
    return 0

# part(b)
def simulate_ttt():
    """
    Simulate one game with random moves.
    Returns: 1 if A wins, -1 if B wins, 0 if draw.
    """
    board = [0] * 9  # Initialize empty board
    current_player = 1  # Player A starts (1)

    while True:
        # Find all empty positions
        empty_positions = []
        for i in range(9):
            if board[i] == 0:
                empty_positions.append(i)

        # If no empty positions, it's a draw
        if len(empty_positions) == 0:
            return 0
        # Randomly choose the empty position
        chosen_pos = random.choice(empty_positions)
        board[chosen_pos] = current_player
        # Check for winner
        winner = check_winner(board) # Check for winner after the actual move
        if winner != 0:
            return winner  # Return 1 or -1 or 0 for draw
        # Switch player
        current_player *= -1

# Main Program for part(b)    
print("Part b: Simulating 1000 random games. Random seed = 3042")
wins_A = 0
wins_B = 0
draws = 0
random.seed(3042)
for _ in range(1000):
    result = simulate_ttt()
    if result == 1:
        wins_A += 1
    elif result == -1:
        wins_B += 1
    else:  # result == 0
        draws += 1

# Print the final results for Part b
print(f"Player A won {wins_A} times.")
print(f"Player B won {wins_B} times.")
print(f"The game had no winner {draws} times.")

# part(c)
def simulate_clever_ttt():
    """
    Players check for winning move first, then blocking move, then random.
    Returns: 1 if A wins, -1 if B wins, 0 if draw.
    """ 
    board = [0] * 9  # Initialize empty board
    current_player = 1  # Player A starts (1)
    while True:
        # Find all empty positions
        empty_positions = []
        for i in range(9):
            if board[i] == 0:
                empty_positions.append(i)
        if len(empty_positions) == 0:
            return 0

        chosen_pos = None

        # Step 1: Check for a winning move for current player
        for pos in empty_positions:
            # Temporarily place current player's mark
            board[pos] = current_player
            # Check if this move wins
            if check_winner(board) == current_player:
                chosen_pos = pos  # Found winning move!
                break
            board[pos] = 0

        # Step 2: If no winning move, check for a blocking move (opponent's winning move)
        if chosen_pos is None:
            for pos in empty_positions:
                # Temporarily place opponent's mark
                board[pos] = current_player * -1
                # Check if this move would let opponent win
                if check_winner(board) == current_player * -1:
                    chosen_pos = pos  # Found blocking move!
                    break
                board[pos] = 0

        # Step 3: If no winning or blocking move, choose randomly
        if chosen_pos is None:
            chosen_pos = random.choice(empty_positions)

        # Actually make the move
        board[chosen_pos] = current_player

        # Check for winner after the actual move
        winner = check_winner(board)
        if winner != 0:
            return winner  # Return 1 or -1

        # Switch player
        current_player *= -1

#Main Program for part(c)
print("\nPart c: Simulating 1000 clever games. Random seed = 42")
wins_A_clever = 0
wins_B_clever = 0
draws_clever = 0
random.seed(42) 

for _ in range(1000):
    result = simulate_clever_ttt()
    if result == 1:
        wins_A_clever += 1
    elif result == -1:
        wins_B_clever += 1
    else:  # result == 0
        draws_clever += 1

# Print the final results for Part c (as required by the assignment)
print(f"Player A won {wins_A_clever} times (clever).")
print(f"Player B won {wins_B_clever} times (clever).")
print(f"The game had no winner {draws_clever} times (clever).")
```

Q2(4 marks)

```python
def simulate_blind_box():
    """
    Simulate one trial of collecting all 6 styles of blind-box toys.
    Returns the number of boxes bought to collect all 6 styles.
    """
    # Define styles and their probabilities
    styles = ['A', 'B', 'C', 'D', 'E', 'F']
    probabilities = [0.05,0.1,0.15,0.2,0.2,0.3]

    # Use a set to track collected styles and a counter for boxes bought
    collected = set()
    boxes_bought = 0

    # Continue buying boxes until all styles are collected
    while len(collected) < 6:
        # Generate a random number between 0 and 1
        rand_num = random.random()
        # Determine which style was drawn based on cumulative probability
        cumulative_prob = 0.0
        for i in range(len(styles)):
            cumulative_prob += probabilities[i]
            if rand_num < cumulative_prob:
                collected.add(styles[i])
                break  # Exit the for loop, go back to while loop

        boxes_bought += 1

    return boxes_bought


# Main Program
print("Simulating 1000 trials of blind-box collection. Random seed = 3042")
total_boxes = 0
num_trials = 1000
random.seed(3042) 
for _ in range(num_trials):
    boxes_needed = simulate_blind_box()
    total_boxes += boxes_needed

average_boxes = (total_boxes / num_trials)

print(f"On average, need to buy {average_boxes} boxes to collect all styles.")
```

Q3
(Part a, 1 marks)
(Part b, 2 marks)
(Part c, 3 marks)

```python
import random

#part(a)
"""
The coordinates (x,y) of any position where a person is allowed to walk satisfy the following:
For horizontal movement: the move is (±1,0)
For vertical movement: the move is (0,±1)
For diagonal movement: the move is (−1,−1) or (+1,+1)
"""

# part(b)
def avaliable_move(x, y):
    """
    Get all possible directions from position (x, y) based on grid rules.
    """
    directions = []
    if y % 3 == 0:
        if x>0:
            directions.append((-1, 0))  # left
        if x<15:
            directions.append((1, 0))   # right
    if x % 3 == 0:
        if y>0:
            directions.append((0, -1))  # down
        if y<15:
            directions.append((0, 1))   # up
    
    # Diagonal moves only allowed when both x and y are multiples of 3
    if abs(y-x) % 3 == 0 and -15 <= (y - x) <= 15:
        if x < 15 and y < 15:
            directions.append((1, 1))   # up-right
        if x > 0 and y > 0:
            directions.append((-1, -1)) # down-left   
    
    # print(directions) 
    return directions

def step(x, y, memory):
    """
    Args:
        x, y: current integer coordinates
        memory: previous position as (prev_x, prev_y) or None
    
    Returns:
        (new_x, new_y, new_memory)
    """
    # Get all possible directions from current position
    all_dirs = avaliable_move(x, y)
    
    # Handle case where no moves are available
    if not all_dirs:
        if memory is not None:
            # Move back to previous position
            prev_x, prev_y = memory
            return prev_x, prev_y, (x, y)
        else:
            # Cannot move anywhere, stay in place
            return x, y, memory
    
    # Remove the direction that goes back to memory (if memory exists)
    valid_dirs = []
    for dx, dy in all_dirs:
        new_x, new_y = x + dx, y + dy
        if memory is not None:
            prev_x, prev_y = memory
            if new_x == prev_x and new_y == prev_y:
                continue
        valid_dirs.append((dx, dy))
    
    # If no valid directions, use all directions (including back to memory)
    if not valid_dirs:
        valid_dirs = all_dirs
    
    # Choose randomly
    dx, dy = random.choice(valid_dirs)
    new_x, new_y = x + dx, y + dy
    new_memory = (x, y)
    
    return new_x, new_y, new_memory


# part(c)
def move_toward(x1, y1, x2, y2):
    """
    Return the direction (dx, dy) for person at (x1, y1) to move 1 step toward (x2, y2)
    along the corridor they share.
    """
    dx, dy = 0, 0
    
    if x1 != x2:
        dx = 1 if x2 > x1 else -1
    if y1 != y2:
        dy = 1 if y2 > y1 else -1
    
    # But we must ensure this move is legal!
    # However, since they are on same corridor, this move should be valid.
    return dx, dy

def can_see_each_other(x1, y1, x2, y2):
    """
    Check if two people can see each other along a corridor.
    
    Conditions:
    1. Diagonal visibility: Points lie on line y = x + b where b ∈ [-15, 15] and b % 3 == 0
    2. Horizontal visibility: y1 == y2 and y1 ∈ [0, 15] and y1 % 3 == 0  
    3. Vertical visibility: x1 == x2 and x1 ∈ [0, 15] and x1 % 3 == 0
        
    Returns:
        bool: True if they can see each other, False otherwise
    """
    # Check diagonal visibility (slope = 1)
    if x1 - y1 == x2 - y2:  # This ensures they are on the same line y = x + b
        b = y1 - x1  # intercept of the line y = x + b
        if -15 <= b <= 15 and b % 3 == 0:
            return True
    
    # Check horizontal visibility
    if y1 == y2:
        if 0 <= y1 <= 15 and y1 % 3 == 0:
            return True
    
    # Check vertical visibility  
    if x1 == x2:
        if 0 <= x1 <= 15 and x1 % 3 == 0:
            return True
    
    return False

def simulate_supermarket():
    """
    Simulate one run of A and B walking until they meet.
    A starts at (15, 0), B at (0, 15).
    """
    # Initial positions
    ax, ay = 15, 0
    bx, by = 0, 15
    
    # Memory: previous positions
    a_mem = None
    b_mem = None
    
    total_steps = 0
    
    while True:
        # Check if they meet
        if ax == bx and ay == by:
            break
        
        # ===== A's turn =====
        if can_see_each_other(ax, ay, bx, by) is True:
            # Move directly toward B (ignore memory)
            dx, dy = move_toward(ax, ay, bx, by)
            ax += dx
            ay += dy
            a_mem = (ax - dx, ay - dy)  # update memory to previous
        else:
            # Random walk with memory
            ax, ay, a_mem = step(ax, ay, a_mem)
        total_steps += 1
        
        # Check meet after A's move
        if ax == bx and ay == by:
            break
        
        # ===== B's turn =====
        if can_see_each_other(bx, by, ax, ay)is True:
            dx, dy = move_toward(bx, by, ax, ay)
            bx += dx
            by += dy
            b_mem = (bx - dx, by - dy)
        else:
            bx, by, b_mem = step(bx, by, b_mem)
        total_steps += 1
    
    return total_steps

if __name__ == "__main__":
    num_simulations = 1000
    all_simulation_steps = []

    print(f"Running {num_simulations} simulations...")
    
    for i in range(num_simulations):
        steps_for_this_run = simulate_supermarket()
        all_simulation_steps.append(steps_for_this_run)

    average_steps = sum(all_simulation_steps) / num_simulations

    print("\n--- Simulation Results ---")
    print(f"Across {num_simulations} simulations,")
    print(f"it took an average of {average_steps:.2f} total steps for the two people to meet.")
```

Q4 (4 marks)

```python
# Case 1: Only one bullet in the gun
import numpy as np
import time

num_test = 10000
num_pos = 6
pos_take = np.array([1, 3, 4]) 

start_time = time.time()

# Generate ALL bullet positions at once
bullet_positions = np.random.randint(0, num_pos, size=num_test)

matches = bullet_positions[:, np.newaxis] == pos_take
lose_mask = np.any(matches, axis=1)

num_lose = np.sum(lose_mask)

lose_probability = num_lose / num_test
end_time = time.time()
total_duration = time.time() - start_time

print("\nCase 1: Only one bullet in the gun")
print("The lose probability is:", lose_probability)
print(f"start_time:{start_time},end_time:{end_time}")
print("Time:", total_duration)
```

```python
# Case 2: Multiple bullets in the gun

import numpy as np
import time

num_test = 10000
num_pos = 10
num_bullets = 3
pos_take = np.array([0, 1, 2, 3, 4])

start_time = time.time()

bullet_positions = np.random.randint(0, num_pos, size=(num_test, num_bullets))
lose_mask = np.isin(bullet_positions, pos_take).any(axis=1)

num_lose = np.sum(lose_mask)
lose_probability = num_lose / num_test
total_duration = time.time() - start_time

print("\nCase 2: Multiple bullets in the gun")
print("The lose probability is:", lose_probability)
print("Time:", total_duration)
```
